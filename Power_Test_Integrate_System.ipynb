````python
# æ¯›ç»ç’ƒæ•ˆæœæŒ‰é’®ç»„ä»¶
class SimpleGlassButton:
    """ç°ä»£åŒ–æ¯›ç»ç’ƒæ•ˆæœæŒ‰é’®ï¼Œé€‚é…æ¸å˜èƒŒæ™¯"""
    
    def __init__(self, parent, x, y, width, height, text, command=None):
        self.parent = parent
        self.x = x
        self.y = y
        self.width = width
        self.height = height
        self.text = text
        self.command = command
        self.hovered = False
        self.clicked = False
        
        # åˆ›å»ºæŒ‰é’®ç”»å¸ƒ
        self.canvas = tk.Canvas(
            parent, 
            width=width, 
            height=height,
            highlightthickness=0,
            bd=0,
            bg=parent['bg'] if 'bg' in parent.keys() else '#87CEEB'
        )
        self.canvas.place(x=x, y=y)
        
        # ç»‘å®šäº‹ä»¶
        self.canvas.bind("<Button-1>", self._on_click)
        self.canvas.bind("<ButtonRelease-1>", self._on_release)
        self.canvas.bind("<Enter>", self._on_enter)
        self.canvas.bind("<Leave>", self._on_leave)
        
        self._draw_button()
    
    def _draw_button(self):
        """ç»˜åˆ¶æ¯›ç»ç’ƒæ•ˆæœæŒ‰é’®"""
        self.canvas.delete("all")
        
        # è®¡ç®—é¢œè‰²çŠ¶æ€
        if self.clicked:
            color = BUTTON_ACTIVE
            opacity = 0.85
        elif self.hovered:
            color = BUTTON_HOVER
            opacity = 0.75
        else:
            color = BUTTON_COLOR
            opacity = 0.65
        
        # ç»˜åˆ¶æ¯›ç»ç’ƒåœ†è§’çŸ©å½¢
        radius = 8
        
        # ä¸»ä½“èƒŒæ™¯
        self.canvas.create_rectangle(
            radius, 0, self.width - radius, self.height,
            fill=color, outline="", width=0
        )
        self.canvas.create_rectangle(
            0, radius, self.width, self.height - radius,
            fill=color, outline="", width=0
        )
        
        # åœ†è§’æ•ˆæœ
        for i in range(radius):
            y1 = i
            y2 = self.height - i - 1
            x_offset = int(math.sqrt(radius*radius - (radius-i)*(radius-i)))
            
            self.canvas.create_line(
                x_offset, y1, self.width - x_offset, y1,
                fill=color, width=1
            )
            self.canvas.create_line(
                x_offset, y2, self.width - x_offset, y2,
                fill=color, width=1
            )
        
        # æ¯›ç»ç’ƒé«˜å…‰æ•ˆæœ
        if not self.clicked:
            highlight_color = '#FFFFFF'
            self.canvas.create_rectangle(
                1, 1, self.width - 1, self.height // 3,
                fill=highlight_color, stipple='gray25', outline=""
            )
        
        # æ–‡å­—
        text_color = '#FFFFFF' if not self.clicked else '#F0F8FF'
        self.canvas.create_text(
            self.width // 2, self.height // 2,
            text=self.text, fill=text_color,
            font=('Microsoft YaHei UI', 10, 'bold')
        )
    
    def _on_click(self, event):
        self.clicked = True
        self._draw_button()
        if self.command:
            self.command()
    
    def _on_release(self, event):
        self.clicked = False
        self._draw_button()
    
    def _on_enter(self, event):
        self.hovered = True
        self._draw_button()
    
    def _on_leave(self, event):
        self.hovered = False
        self._draw_button()

print("ğŸ¨ æ¯›ç»ç’ƒæŒ‰é’®ç»„ä»¶å·²å®šä¹‰")
````

````python
# æ¸å˜èƒŒæ™¯ç»˜åˆ¶å‡½æ•°
def draw_gradient_background(canvas, width, height, start_color, end_color):
    """ç»˜åˆ¶å‚ç›´æ¸å˜èƒŒæ™¯ (ä¸Šæµ…ä¸‹æ·±)"""
    canvas.delete("gradient")  # æ¸…é™¤ä¹‹å‰çš„æ¸å˜
    
    # é€è¡Œç»˜åˆ¶æ¸å˜æ•ˆæœ
    for i in range(height):
        # è®¡ç®—å½“å‰è¡Œçš„æ··åˆæ¯”ä¾‹
        ratio = i / height
        
        # RGBé¢œè‰²æ’å€¼è®¡ç®—
        r = int(start_color[0] * (1 - ratio) + end_color[0] * ratio)
        g = int(start_color[1] * (1 - ratio) + end_color[1] * ratio)
        b = int(start_color[2] * (1 - ratio) + end_color[2] * ratio)
        
        # è½¬æ¢ä¸º16è¿›åˆ¶é¢œè‰²
        color = f"#{r:02x}{g:02x}{b:02x}"
        
        # ç»˜åˆ¶æ°´å¹³çº¿
        canvas.create_line(
            0, i, width, i, 
            fill=color, width=1, 
            tags="gradient"
        )
    
    print(f"ğŸ¨ æ¸å˜èƒŒæ™¯å·²ç»˜åˆ¶ ({width}x{height})")

# é¢œè‰²è½¬æ¢å·¥å…·å‡½æ•°
def hex_to_rgb(hex_color):
    """å°†16è¿›åˆ¶é¢œè‰²è½¬æ¢ä¸ºRGBå…ƒç»„"""
    hex_color = hex_color.lstrip('#')
    return tuple(int(hex_color[i:i+2], 16) for i in (0, 2, 4))

def rgb_to_hex(rgb_color):
    """å°†RGBå…ƒç»„è½¬æ¢ä¸º16è¿›åˆ¶é¢œè‰²"""
    return f"#{rgb_color[0]:02x}{rgb_color[1]:02x}{rgb_color[2]:02x}"

print("ğŸŒˆ æ¸å˜èƒŒæ™¯ç»˜åˆ¶å‡½æ•°å·²å®šä¹‰")
````

````python
# ä¸»ç•Œé¢ç±»
class SimpleTkinterToolsInterface:
    """ç®€åŒ–çš„ tkintertools é£æ ¼ä¸»ç•Œé¢"""
    
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Power Test Integrate System - Notebook Version")
        self.root.geometry(f"{WINDOW_WIDTH}x{WINDOW_HEIGHT}")
        self.root.resizable(False, False)
        
        # è®¾ç½®çª—å£å›¾æ ‡å’Œæ ·å¼
        try:
            self.root.iconbitmap('icon.ico') if os.path.exists('icon.ico') else None
        except:
            pass
        
        # åˆå§‹åŒ–ç•Œé¢
        self._setup_interface()
        print("ğŸ–¥ï¸ ä¸»ç•Œé¢åˆå§‹åŒ–å®Œæˆ")
    
    def _setup_interface(self):
        """è®¾ç½®ç•Œé¢å¸ƒå±€"""
        # ä¸»ç”»å¸ƒ (ç”¨äºæ¸å˜èƒŒæ™¯)
        self.main_canvas = tk.Canvas(
            self.root,
            width=WINDOW_WIDTH,
            height=WINDOW_HEIGHT,
            highlightthickness=0,
            bd=0
        )
        self.main_canvas.pack(fill="both", expand=True)
        
        # ç»˜åˆ¶æ¸å˜èƒŒæ™¯
        draw_gradient_background(
            self.main_canvas, 
            WINDOW_WIDTH, 
            WINDOW_HEIGHT,
            GRADIENT_START, 
            GRADIENT_END
        )
        
        # åˆ›å»ºç•Œé¢å…ƒç´ 
        self._create_header()
        self._create_buttons()
        self._create_status_area()
    
    def _create_header(self):
        """åˆ›å»ºé¡¶éƒ¨æ ‡é¢˜åŒºåŸŸ"""
        # Logo å’Œæ ‡é¢˜
        self.main_canvas.create_text(
            50, 30,
            text="âš¡ Power Test System",
            font=('Microsoft YaHei UI', 16, 'bold'),
            fill='white',
            anchor='w'
        )
        
        # ç‰ˆæœ¬ä¿¡æ¯
        self.main_canvas.create_text(
            50, 55,
            text="Notebook Interactive Version v2.0",
            font=('Microsoft YaHei UI', 10),
            fill='#E6F3FF',
            anchor='w'
        )
    
    def _create_buttons(self):
        """åˆ›å»ºåŠŸèƒ½æŒ‰é’®"""
        # æŒ‰é’®ä½ç½®é…ç½®
        button_configs = [
            (300, 25, "ğŸ“Š æ•°æ®é‡‡é›†"),
            (440, 25, "ğŸ”§ è®¾å¤‡æ§åˆ¶"),
            (580, 25, "ğŸ“ˆ å®æ—¶ç›‘æ§"),
            (720, 25, "ğŸ’¾ æ•°æ®å¯¼å‡º"),
            (860, 25, "âš™ï¸ ç³»ç»Ÿè®¾ç½®"),
            (1000, 25, "â“ å¸®åŠ©")
        ]
        
        self.buttons = {}
        for x, y, text in button_configs:
            btn_name = text.split()[-1]  # æå–æŒ‰é’®åç§°
            self.buttons[btn_name] = SimpleGlassButton(
                self.main_canvas,
                x, y, BUTTON_WIDTH, BUTTON_HEIGHT,
                text,
                command=lambda name=btn_name: self._button_clicked(name)
            )
    
    def _create_status_area(self):
        """åˆ›å»ºçŠ¶æ€æ˜¾ç¤ºåŒºåŸŸ"""
        # çŠ¶æ€ä¿¡æ¯èƒŒæ™¯
        status_y = 100
        status_height = WINDOW_HEIGHT - 120
        
        # åŠé€æ˜çŠ¶æ€åŒºåŸŸ
        self.main_canvas.create_rectangle(
            20, status_y, WINDOW_WIDTH - 20, status_y + status_height,
            fill='#FFFFFF', stipple='gray25', outline='#FFFFFF', width=1
        )
        
        # çŠ¶æ€æ ‡é¢˜
        self.main_canvas.create_text(
            40, status_y + 20,
            text="ğŸ“‹ ç³»ç»ŸçŠ¶æ€",
            font=('Microsoft YaHei UI', 12, 'bold'),
            fill='white',
            anchor='w'
        )
        
        # ç³»ç»Ÿä¿¡æ¯
        info_lines = [
            f"ğŸ•’ ç³»ç»Ÿæ—¶é—´: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}",
            "ğŸ”Œ è®¾å¤‡çŠ¶æ€: å·²è¿æ¥",
            "ğŸ“Š æ•°æ®é‡‡é›†: å°±ç»ª",
            "ğŸ’¾ å­˜å‚¨ç©ºé—´: å……è¶³",
            "ğŸŒ ç½‘ç»œè¿æ¥: æ­£å¸¸"
        ]
        
        for i, line in enumerate(info_lines):
            self.main_canvas.create_text(
                40, status_y + 50 + i * 25,
                text=line,
                font=('Microsoft YaHei UI', 10),
                fill='#F0F8FF',
                anchor='w'
            )
    
    def _button_clicked(self, button_name):
        """æŒ‰é’®ç‚¹å‡»äº‹ä»¶å¤„ç†"""
        timestamp = datetime.datetime.now().strftime('%H:%M:%S')
        print(f"[{timestamp}] ğŸ”˜ æŒ‰é’®ç‚¹å‡»: {button_name}")
        
        # æ›´æ–°çŠ¶æ€æ˜¾ç¤º
        self.main_canvas.create_text(
            40, WINDOW_HEIGHT - 50,
            text=f"æœ€è¿‘æ“ä½œ: {button_name} ({timestamp})",
            font=('Microsoft YaHei UI', 10),
            fill='#FFFF99',
            anchor='w',
            tags="last_action"
        )
        
        # æ¸…é™¤ä¹‹å‰çš„æ“ä½œè®°å½•
        self.main_canvas.delete("last_action")
        self.main_canvas.create_text(
            40, WINDOW_HEIGHT - 50,
            text=f"æœ€è¿‘æ“ä½œ: {button_name} ({timestamp})",
            font=('Microsoft YaHei UI', 10),
            fill='#FFFF99',
            anchor='w',
            tags="last_action"
        )
    
    def run(self):
        """å¯åŠ¨ç•Œé¢"""
        print("ğŸš€ å¯åŠ¨ Power Test Integrate System...")
        self.root.mainloop()

print("ğŸ—ï¸ ä¸»ç•Œé¢ç±»å·²å®šä¹‰")
````

````python
# é”™è¯¯å¤„ç†å’Œè°ƒè¯•å·¥å…·
class SystemDebugger:
    """ç³»ç»Ÿè°ƒè¯•å’Œé”™è¯¯å¤„ç†å·¥å…·"""
    
    @staticmethod
    def check_dependencies():
        """æ£€æŸ¥ç³»ç»Ÿä¾èµ–"""
        print("ğŸ” æ£€æŸ¥ç³»ç»Ÿä¾èµ–...")
        deps = {
            'tkinter': True,
            'datetime': True,
            'math': True,
            'sys': True,
            'os': True
        }
        
        for dep, status in deps.items():
            status_icon = "âœ…" if status else "âŒ"
            print(f"  {status_icon} {dep}: {'å¯ç”¨' if status else 'ä¸å¯ç”¨'}")
        
        return all(deps.values())
    
    @staticmethod
    def test_color_system():
        """æµ‹è¯•é¢œè‰²ç³»ç»Ÿ"""
        print("ğŸ¨ æµ‹è¯•é¢œè‰²ç³»ç»Ÿ...")
        
        try:
            # æµ‹è¯•RGBè½¬æ¢
            test_rgb = (135, 206, 235)
            test_hex = rgb_to_hex(test_rgb)
            back_rgb = hex_to_rgb(test_hex)
            
            print(f"  RGB: {test_rgb} â†’ HEX: {test_hex} â†’ RGB: {back_rgb}")
            
            if test_rgb == back_rgb:
                print("  âœ… é¢œè‰²è½¬æ¢ç³»ç»Ÿæ­£å¸¸")
                return True
            else:
                print("  âŒ é¢œè‰²è½¬æ¢ç³»ç»Ÿå¼‚å¸¸")
                return False
        except Exception as e:
            print(f"  âŒ é¢œè‰²ç³»ç»Ÿé”™è¯¯: {e}")
            return False
    
    @staticmethod
    def validate_window_size():
        """éªŒè¯çª—å£å°ºå¯¸"""
        print("ğŸ“ éªŒè¯çª—å£å°ºå¯¸...")
        
        if WINDOW_WIDTH > 0 and WINDOW_HEIGHT > 0:
            print(f"  âœ… çª—å£å°ºå¯¸: {WINDOW_WIDTH}x{WINDOW_HEIGHT}")
            return True
        else:
            print(f"  âŒ æ— æ•ˆçª—å£å°ºå¯¸: {WINDOW_WIDTH}x{WINDOW_HEIGHT}")
            return False
    
    @staticmethod
    def system_check():
        """å®Œæ•´ç³»ç»Ÿæ£€æŸ¥"""
        print("ğŸ”§ æ‰§è¡Œç³»ç»Ÿæ£€æŸ¥...")
        checks = [
            SystemDebugger.check_dependencies(),
            SystemDebugger.test_color_system(),
            SystemDebugger.validate_window_size()
        ]
        
        success_count = sum(checks)
        total_count = len(checks)
        
        if success_count == total_count:
            print(f"âœ… ç³»ç»Ÿæ£€æŸ¥é€šè¿‡ ({success_count}/{total_count})")
            return True
        else:
            print(f"âš ï¸ ç³»ç»Ÿæ£€æŸ¥éƒ¨åˆ†å¤±è´¥ ({success_count}/{total_count})")
            return False

# å¼‚å¸¸å¤„ç†è£…é¥°å™¨
def safe_execute(func):
    """å®‰å…¨æ‰§è¡Œè£…é¥°å™¨ï¼Œæ•è·å¹¶æŠ¥å‘Šå¼‚å¸¸"""
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            print(f"âŒ æ‰§è¡Œé”™è¯¯ in {func.__name__}: {e}")
            return None
    return wrapper

print("ğŸ› ï¸ è°ƒè¯•å·¥å…·å·²å®šä¹‰")
````
